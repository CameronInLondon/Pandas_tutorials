Following from my last blog on Selecting columns and Filtering rows in Pandas, I continue this series on the Python Pandas library for data analysis. Today I wanted to talk about shaping data or transforming data.

Data has a habit of not being ready to use when you receive it. This is why data people tend to spend more then 50% of their time cleaning and shaping data.


I have spoken about tidy data in the past. Tidy data is what data people want to work with but to get the dataset into the shape you want transforming might be needed. Pandas makes it easy by offering some handy dataframe methods.

In this post I will look at pivot(), melt(), stack() and unstack()

![](2021-12-30-15-08-15.png)

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

Lets load some data from Seaborn
https://seaborn.pydata.org/generated/seaborn.load_dataset.html

```python
# sns.get_dataset_names()
flights = sns.load_dataset('flights')
flights.head()
```

## Pivot

Pivot changes data from tall (tidy) format to wide format, like this:-

```python
flights = flights.pivot(columns='month', index='year', values='passengers')
flights.head(3)
```

To get the dataframe back to were we started we can used stack() with reset_index() which will be discussed shortly. 

## Melt

Melt() allows you to go from wide to tall data. Lets load the flights dateset again.

```python
flights_wide = pd.read_csv('flights.csv')
print(flights_wide.shape)
flights_wide.head(3)
```

Now we melt the dataset from wide to long. It goes from 12 to 144 rows long.

```python
flights_long = pd.melt(flights_wide, id_vars='year', var_name='month', value_name='passengers_new').sort_values(by='year').reset_index(drop=True)
print(flights_long.shape)
flights_long.head()
```

To get back to wide data as you know we can just use the pivot method again.

```python
flights_wide = flights_long.pivot(index='year', columns='month', values='passengers_new')
flights_wide.head(3)
```

## Stack

Stack takes the column index and stacks it in the row index. This lengthens the dataset from 12 to 156 rows.

```python
print(f'flights_wide shape is: {flights_wide.shape}')
stack_flights = flights_wide.stack()
stack_flights
```

As an aside, we can then reset the index to get the data back into columns and rename the values column by chaining dataframe methods.

```python
flights_wide.stack().reset_index().rename(columns={0:'values'}).head()
```

## Unstack

Unstack does the reverse of stack.

```python
stack_flights.unstack().head(3)
```

## Transposing

It is quick and easy to transpose your dataset in Pandas, there is even a shortcut!

```python
flights.head(3)
```

```python
flights.transpose().head(3)
```
